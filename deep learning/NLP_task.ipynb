```python
import requests
content = requests.get("https://www.gutenberg.org/files/11/11-0.txt")
```

```python
text=content.text
text
```

**LowerCase Data**

```python
text=text.lower()
text
```

**Removing** **Punctuation**

```python
import re
text = re.sub(r'\s+', ' ', text)
text = text.strip()
text
```

```python
text = re.sub(r'\*+', '', text)
text = re.sub(r'\[.*?\]', '', text)
text = re.sub(r' {2,}', ' ', text)
text = text.strip()
text
```

```python
import spacy

# Load the English tokenizer
nlp = spacy.load('en_core_web_sm')
# Tokenize the text
doc = nlp(text)

# Extract tokens
tokens = [token.text for token in doc]

print("Word Tokens:")
print(tokens)
```

**stopwords**

```python
# Remove stopwords
tokens_without_stopwords = [token.text for token in doc if not token.is_stop]

print("Tokens without Stopwords:")
print(tokens_without_stopwords)
```

**Lemmatization**

```python
lemmatized_tokens = [token.lemma_ for token in doc]

print("Lemmatized Tokens:")
print(lemmatized_tokens)
```

**Bag of Words (BoW)**

```python
from sklearn.feature_extraction.text import CountVectorizer

texts = [text]
# texts = [
#     "Alice was beginning to get very tired",
#     "She had peeped into the book her sister was reading"
# ]

vectorizer = CountVectorizer()


X = vectorizer.fit_transform(texts)

# Convert the result into an array and display the feature names
print("BoW Model (Features):")
print(vectorizer.get_feature_names_out())
print("Vectorized Data:")
print(X.toarray())

```

**TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample text data (after preprocessing)
texts = [text]

# Initialize TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data into a vector
X_tfidf = tfidf_vectorizer.fit_transform(texts)

# Convert the result into an array and display the feature names
print("TF-IDF Model (Features):")
print(tfidf_vectorizer.get_feature_names_out())
print("TF-IDF Vectorized Data:")
print(X_tfidf.toarray())
```

**Word2Vec (Word Embeddings)**

```python
import gensim
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import word_tokenize

# Sample text data (after preprocessing)
text = "Alice was beginning to get very tired of sitting by her sister."

# Tokenize the text
tokens = word_tokenize(text)

# Train a Word2Vec model
model = Word2Vec([tokens], min_count=1)

# Get the vector representation for a word
word_vector = model.wv['alice']

print("Word2Vec Vector for 'alice':")
print(word_vector)

```

```python
def GetPage(Link,FileName) :
    page = requests.get(Link)
    soup = BeautifulSoup(page.content, 'html.parser')
    print(f'Number of paragraphs is :  {len(soup.find_all("p"))}')
    try :
        title = soup.find(id="firstHeading")
        print(f'Page title is :   {title.string}')
    except :
        pass
    if len(soup.find_all('p')) ==0 : return None
    f = open(FileName,'w',encoding = 'utf-8')
    for i in range(len(soup.find_all('p'))) :
        f.write(soup.find_all('p')[i].get_text())
        f.write('\n')
    f.close()
```
